# AI Rosary Habit Companion
**An 80-Day Spiritual Transformation Agent**

## 🚀 The Mission
Consistency is the hardest part of building a spiritual habit. This project moves beyond static reminders by architecting an automated system that integrates daily prayer into a digital workflow. It is designed to support an 80-day Rosary challenge through proactive scheduling and data ownership.

## 🛠️ The Tech Stack
- **Python & Streamlit**: A mobile-friendly dashboard for daily "check-ins."
- **Google Calendar API**: Proactive integration that syncs the 80-day plan with my personal schedule.
- **SQLite (The Habit Ledger)**: A local database that tracks every "Completed" day, ensuring 100% data privacy and ownership.
- **Liturgical Logic**: Automated identification of the daily Mystery (Joyful, Sorrowful, Glorious, Luminous) based on the day of the week.

## 📈 Key Features
- **Smart Scheduling**: Generates an 80-day calendar series with 20-minute time blocks.
- **Progress Tracking**: A live progress bar that visualizes the journey from Day 1 to Day 80.
- **The "Success" Loop**: An interactive "Mark Completed" button that triggers celebratory UI elements (balloons) and updates the permanent ledger.

```python
import os.path
import datetime as dt
import pandas as pd
import streamlit as st
import pickle
from google.auth.transport.requests import Request
from google.oauth2.credentials import Credentials
from google_auth_oauthlib.flow import InstalledAppFlow
from googleapiclient.discovery import build
import sqlite3

# The corrected "Value Creator" Scope with the required suffix
SCOPES = ["https://www.googleapis.com/auth/calendar"]

print(f"Current Scope: {SCOPES}") # This proves to us it's correct!
```

```python
#First hand: Authentication function
def get_calendar_services():
    creds = None
    #file token.pickle stores the user's access and refresh tokens
    if os.path.exists("token.pickle"):
        with open("token.pickle","rb") as token:
            creds = pickle.load(token)

    if not creds or not creds.valid:
        if creds and creds.expired and creds.refresh_token:
            creds.refresh(Request())
        else:
            flow = InstalledAppFlow.from_client_secrets_file("credentials.json", SCOPES)
            creds = flow.run_local_server(port=0)
        with open("token.pickle","wb") as token:
            pickle.dump(creds,token)

    #calender was changed to calendar
    return build("calendar","v3", credentials=creds)

#Second hand: For the Reminder Creation Function
service = get_calendar_services()
def create_80_day_challenge(service, start_date_str):
    # 1. Convert start string to a datetime object
    start_date = dt.datetime.fromisoformat(start_date_str)
    
    print(f"🚀 Starting 80-Day Rosary Challenge from {start_date.date()}...")

    for day in range(1, 81):
        # Calculate the date for this specific day of the challenge
        current_date = start_date + dt.timedelta(days=day-1)
        
        # Determine the Mystery based on the day of the week
        day_name = current_date.strftime("%A")
        mysteries = {
            "Monday": "Joyful", "Tuesday": "Sorrowful", "Wednesday": "Glorious",
            "Thursday": "Luminous", "Friday": "Sorrowful", "Saturday": "Joyful",
            "Sunday": "Glorious"
        }
        mystery = mysteries.get(day_name)

        # Define the Event structure
        event_body = {
            'summary': f'Rosary Day {day}: {mystery} Mysteries',
            'description': f'Day {day} of your 80-day habit transformation. Meditation on the {mystery} Mysteries.',
            'start': {
                'dateTime': current_date.isoformat() + 'Z',
                'timeZone': 'UTC',
            },
            'end': {
                'dateTime': (current_date + dt.timedelta(minutes=40)).isoformat() + 'Z',
                'timeZone': 'UTC',
            },
            'colorId': '11', # Google's "Bold Blue" to make it stand out
            'reminders': {
                'useDefault': False,
                'overrides': [{'method': 'popup', 'minutes': 15}],
            },
        }

        # 2. Push to Google Calendar
        # 3. Push to your mobile-synced primary calendar
        result = service.events().insert(calendarId='primary', body=event_body).execute()
        print(f"✅ 80-Day Automation Active! Link: {result.get('htmlLink')}")
        
        if day % 10 == 0 or day == 1:
            print(f"✅ Day {day} synced! Mystery:{mystery}")
            
    print("✨ Mission Accomplished: All 80 days are now on your calendar!")

create_80_day_challenge(service, "2026-04-05T20:00:00")
#For Deleting all the events in the calendar app
#service.calendars().clear(calendarId='primary').execute()
```

```python
def cleanup_duplicate_rosaries(service):
    # 1. Get all events from now onwards
    now = dt.datetime.utcnow().isoformat() + 'Z'
    events_result = service.events().list(calendarId='primary', timeMin=now,
                                          singleEvents=True, orderBy='startTime').execute()
    events = events_result.get('items', [])

    # 2. Track what we've seen to find duplicates
    seen_summaries = {}
    deleted_count = 0

    for event in events:
        summary = event.get('summary', '')
        if "📿 Rosary Day" in summary:
            # If we've already seen this Day X, delete this copy
            if summary in seen_summaries:
                service.events().delete(calendarId='primary', eventId=event['id']).execute()
                deleted_count += 1
            else:
                seen_summaries[summary] = True

    print(f"🧹 Cleanup complete! Deleted {deleted_count} duplicate events.")

# Run the cleanup
cleanup_duplicate_rosaries(service)
```

```python
def initialize_habit_ledger(start_date_str):
    # 1. Connect to the database
    conn = sqlite3.connect('rosary_tracker.db')
    cursor = conn.cursor()

    # 2. Create the Ledger Table
    cursor.execute('''
        CREATE TABLE IF NOT EXISTS habit_ledger (
            day_number INTEGER PRIMARY KEY,
            date TEXT,
            mystery TEXT,
            status TEXT DEFAULT 'Pending',
            completion_time TEXT
        )
    ''')

    # 3. Populate 80 days if the ledger is empty
    cursor.execute("SELECT COUNT(*) FROM habit_ledger")
    if cursor.fetchone()[0] == 0:
        start_date = dt.datetime.fromisoformat(start_date_str)
        for day in range(1, 81):
            current_date = start_date + dt.timedelta(days=day-1)
            day_name = current_date.strftime("%A")
            
            # Traditional Mystery Logic
            mysteries = {"Monday": "Joyful", "Tuesday": "Sorrowful", "Wednesday": "Glorious",
                         "Thursday": "Luminous", "Friday": "Sorrowful", "Saturday": "Joyful", "Sunday": "Glorious"}
            mystery = mysteries.get(day_name)

            cursor.execute("INSERT INTO habit_ledger (day_number, date, mystery) VALUES (?, ?, ?)", 
                           (day, current_date.date().isoformat(), mystery))
        
        conn.commit()
        print("✅ Success: Your 80-Day Habit Ledger is initialized and ready.")
    else:
        print("ℹ️ Ledger already exists. Ready for updates!")
    
    conn.close()

# Start the ledger from today
initialize_habit_ledger('2026-04-05T20:00:00')
```

```python
#For Marking Done Days Manually
def mark_day_as_done(day_number):
    conn = sqlite3.connect('rosary_tracker.db')
    cursor = conn.cursor()
    
    timestamp = dt.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    
    # Update the status for the specific day
    cursor.execute('''
        UPDATE habit_ledger 
        SET status = 'Completed', completion_time = ? 
        WHERE day_number = ?
    ''', (timestamp, day_number))
    
    conn.commit()
    conn.close()
    print(f"🌟 Day {day_number} marked as COMPLETED. Your streak has begun!")
```

```python
#For Maunally resetting the ledger date
def reset_day(num):
    conn = sqlite3.connect("rosary_tracker.db")
    cursor = conn.cursor()
    cursor.execute("UPDATE habit_ledger SET status='Pending', completion_time = NULL WHERE day_number = ?", (num,))
    conn.commit()
    conn.close()
    print(f"Day {num} has been reset to pending in the dataset memory")
```

```python
#For maunually viewing ledger
def view_ledger():
    conn = sqlite3.connect('rosary_tracker.db')
    df = pd.read_sql_query("SELECT * FROM habit_ledger LIMIT 20", conn)
    conn.close()
    return df

# Look at your first 5 days to see Day 1 marked 'Completed'
print(view_ledger())
```

```python
%%writefile rosary_app.py
#For the streamlit experience
import streamlit as st
import sqlite3
import pandas as pd
import datetime

#Streamlit Page Configuration
st.set_page_config(page_title="Catholic AI Rosary Companion")

# Background coloring
slideshow_css = """
<style>
/*1 Targeting very bootom layer for the animation */
.stApp {
    animation: colorChange 30s infinite alternate !important;
}
/* 2Making all top layer transparent so we can see bottom layer */
[data-testid="stHeader"],
[data-testid="stAppViewContainer"],
[data-testid="stCanvas"],
.main {
    background-color: transparent !important;
    background: transparent !important;
}

@keyframes colorChange {
    20%  { background-color: #2e7d32; }/*Green */
    25%  { background-color: #c62828; }/*Red */
    50%  { background-color: #ffffff; }/*White */
    75%  { background-color: #1565c0; }/*Blue */
    100%  { background-color: #fdd835; }/*Yellow */
}

/* Optional: This is for making text to be readable on changing color background */
h1, h2, h3, p, spam, label {
    color: #1e1e1e !important;
}
</style>
"""
st.markdown(slideshow_css, unsafe_allow_html=True)
st.title("Catholic AI Rosary Habit Builder")
st.markdown("80-Day Spritual Transformation Challenge")

def get_db_connection():
    return sqlite3.connect("rosary_tracker.db", check_same_thread=False, timeout=10)

def get_progress():
    conn = get_db_connection()
    df = pd.read_sql_query("SELECT * FROM habit_ledger", conn)
    conn.close()
    return df
    
def mark_done(day):
    conn = get_db_connection()
    cursor = conn.cursor()
    now = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    cursor.execute("UPDATE habit_ledger SET status = 'Completed', completion_time = ? WHERE day_number = ?", (now, day))
    conn.commit()
    conn.close()

# 3.Load Data & Today's Info
df = get_progress()
today_obj = datetime.date.today()
today_str = today_obj.isoformat()
today_row = df[df["date"] == today_str]

# # 4. Today's Mystery & Action
# today_date = datetime.date.today().isoformat()
# today_data = df[df['date'] == today_date]

st.write(f"### Today: {today_obj.strftime("%A, %B %d, %Y")}")

st.sidebar.header("Your Journey")
df = get_progress()
completed = len(df[df["status"] == "Completed"])
st.sidebar.metric("Days Completed",f"{completed}/80")
st.sidebar.progress(completed / 80)

if not today_row.empty:
    day_num = int(today_row.iloc[0]['day_number'])
    mystery = today_row.iloc[0]['mystery']
    status = today_row.iloc[0]['status']
    
    st.subheader(f"Day {day_num} : {mystery} Mysteries")
    
    if status == 'Pending':
        if st.button(f"Mark Day {day_num} as Completed", use_container_width=True,type="primary"):
            mark_done(day_num)
            st.balloons()
            st.rerun()
        
    else:
        st.button("✅ COMPLETED", disabled=True,use_container_width=True)
        st.success(f"✅ Day {day_num} is already finished. See you tomorrow!")
        
# 5. History Table (Optional)
with st.expander("View Full 80-Day Ledger", expanded=True):
    fresh_df = get_progress()
    st.dataframe(fresh_df[["day_number","date","mystery","status"]], use_container_width=True)
```
